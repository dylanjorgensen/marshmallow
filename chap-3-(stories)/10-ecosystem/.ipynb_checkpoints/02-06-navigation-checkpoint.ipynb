# Attention
- How do we identify what directory has our current attention?

### Terminal
- Max OSX is built on top of the UNIX OS

```python
# Prints our working directory
# $ pwd
```

```python
# Lists the items in our working directory
# $ ls
```

### Python
- These of these as the folder location python's attention is on.

```python
import os
```

```python
# Current working directory
os.getcwd()
```

```python
# Eliminating any symbolic links encountered in the path
os.path.realpath(os.getcwd())
```

### Python

```python
import os
```

```python
# Path to our current working directory
os.getcwd()
```
